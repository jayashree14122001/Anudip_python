```python
'''Write a Pandas program to create a Pivot table and find the total sale amount region wise, manager wise, sales man wise.'''
import pandas as pd     #importing pandas library as pd
import numpy as np      #importing numpy library as np
```

```python
#read the csv file into the DataFrame
df = pd.read_csv("salesdata.csv")     #csv file named "salesdata.csv" is saved in the folder
df.head()
```

```python
#creating a Pivot table and finding the total sale amount region,manager and sales man wise
round(pd.pivot_table(df,index=['Region','Manager','SalesMan'],values=['Sale_amt'],aggfunc=['sum']))
```
