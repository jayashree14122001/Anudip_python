Write a python program using Interpolation to fill in missing values in the data frame. Then generate a subject wise highest score in the form of a bar chart.

```python
import pandas as pd               #importing pandas library as pd
import matplotlib.pyplot as plt   #importing matplotlib.pyplot as plt
import numpy as np                #importing numpy library as numpy
```

```python
data = {'Subject': ['Math', 'Physics', 'Chemistry', 'Biology'],'Student_A': [80, 85, np.nan, 70],'Student_B': [90, np.nan, 75, 85],
        'Student_C': [80, 88, np.nan, 78]}     #given dictionary
df=pd.DataFrame(data)                          #creating a dataframe named 'df' from the above dictionary 'data'
```

```python
df_interpolated= df.interpolate(method="linear")                         #interpolate the dataframe named df

subject_wise_highest=df_interpolated.drop('Subject',axis=1).max(axis=1)  #fetching the highest values of every subjects
plt.figure(figsize=(5,2))                                                #defining the figure size of the bar
plt.bar(df_interpolated['Subject'],subject_wise_highest,color='pink')    #generate a bar chart of subject wise highest scores
plt.title('Subject Wise Highest Scores')                                 #giving a title to the chart
plt.xlabel('Subjects-->')                                                #giving label for x axis
plt.ylabel('Scores-->')                                                  #giving label for y axis
plt.show()                                                               #showing the created bar chart
```

Out of all subject, maths has the highest score and chemistry having the lowest
